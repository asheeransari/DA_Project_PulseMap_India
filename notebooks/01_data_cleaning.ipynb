```python
import numpy as np
import pandas as pd
from thefuzz import process
```

```python
# Load Data
df = pd.read_excel("../data/raw/selected_indicators_data.xlsx")
```

```python
# Basic Check
print("Shape : \n",df.shape)
print("\n Columns : \n",df.columns.tolist())
```

```python
print("Null Values : \n",df.isnull().sum())
```

```python
print("Data Types : \n",df.dtypes)
```

```python
# Statistical summary
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
print(df.describe().round(2).T)
```

```python
# Column rename
df.columns = [
    'district', 'state',
    'bmi_below_normal_w', 'overweight_obese_w',
    'anaemia_w',
    'blood_sugar_w', 'blood_sugar_m',
    'bp_elevated_w', 'bp_elevated_m',
    'cervical_screening_w', 'breast_exam_w', 'oral_exam_w',
    'tobacco_w', 'tobacco_m',
    'alcohol_w', 'alcohol_m',
    'literacy_w',
    'health_insurance',
    'clean_water',
    'sanitation',
    'clean_fuel'
]

print(df.columns.tolist())
print("Shape:", df.shape)
```

```python
# District and State name cleaning
df['district'] = df['district'].str.strip().str.title()
df['state'] = df['state'].str.strip().str.title()

# Check
print("Sample districts:")
print(df['district'].head(10).tolist())
print("\nUnique states:", df['state'].nunique())
print(df['state'].unique())
```

```python
# Numeric columns values 0-100 range check
numeric_cols = df.columns[2:]

for col in numeric_cols:
    below = (df[col] < 0).sum()
    above = (df[col] > 100).sum()
    if below > 0 or above > 0:
        print(f"{col}: {below} below 0, {above} above 100")

print("Range check complete")
```

```python
# Duplicate districts check
duplicates = df[df.duplicated(subset=['district', 'state'], keep=False)]
print(f"Duplicate rows: {len(duplicates)}")
if len(duplicates) > 0:
    print(duplicates[['district', 'state']])
```

```python
# Save cleaned data
df.to_excel('../data/processed/pulsemap_cleaned.xlsx', index=False)
print("Cleaned data saved successfully")
print("Shape:", df.shape)
```
